```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/Cognifyz/Dataset.csv')
data
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Aggregate rating'], bins=30, kde=True)
plt.title('Distribution of Ratings (Histogram)')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Aggregate rating'])
plt.title('Distribution of Ratings (Box Plot)')
plt.xlabel('Aggregate Rating')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x=data['Aggregate rating'])
plt.title('Distribution of Ratings (Violin Plot)')
plt.xlabel('Aggregate Rating')
plt.show()
```

```python
cuisine_avg_rating = data.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)[:10]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=cuisine_avg_rating.index, y=cuisine_avg_rating)
plt.title('Top 10 Cuisines by Average Rating')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
city_avg_rating = data.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)[:10]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=city_avg_rating.index, y=city_avg_rating)
plt.title('Top 10 Cities by Average Rating')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Average Cost for two', y='Aggregate rating', data=data)
plt.title('Average Cost for Two vs. Aggregate Rating')
plt.xlabel('Average Cost for Two')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Votes', y='Aggregate rating', data=data)
plt.title('Votes vs. Aggregate Rating')
plt.xlabel('Votes')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
city_restaurant_count = data['City'].value_counts()[:10]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=city_restaurant_count.index, y=city_restaurant_count)
plt.title('Top 10 Cities by Restaurant Count')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

```python
online_delivery_count = data['Has Online delivery'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(online_delivery_count, labels=['No Online Delivery', 'Online Delivery'], autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Restaurant Online Delivery Availability')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
table_booking_count = data['Has Table booking'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(table_booking_count, labels=['No Table Booking', 'Table Booking'], autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Restaurant Table Booking Availability')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
